```python
# to concatenate  two dataframe

# to create dataframe itself of two country

#to create dataframe of india 

import pandas as pd
indiaweather=pd.DataFrame({
    'city':['mumbai','delhi','jaipur'],
    'temperture':[32,45,60],
    'humidity':[83,60,75]
})
indiaweather
```

```python
#to create dataframe of america 
america_weather=pd.DataFrame({
    'city':['newyork','paris','chicago'],
    'temperture':[72,57,80],
    'humidity':[93,61,98]
})
america_weather
```

```python
#to concatenate  or two dataframe
#create dataframe object as df
df=pd.concat([indiaweather,america_weather])
df
```

```python
#to arrange  index as continue index
df=pd.concat([indiaweather,america_weather],ignore_index=True)
df
```

```python
#to divide dataframe into two frame
newdf=pd.concat([indiaweather,america_weather],keys=["india","USA"])
newdf
```
