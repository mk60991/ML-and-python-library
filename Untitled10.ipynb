```python
#to scatter plot data for 1d data
import matplotlib.pyplot as plt
x=[12,23,45,67,89,90]
y=[11,22,33,44,55,66]

#s=size, marker=shown type as plot such as >,<,o,circle......
plt.scatter(x,y,label="1st scatter",color="red",s=210,marker=">")
plt.xlabel("xdata")
plt.title("scatter cahrt")
plt.ylabel("ydata")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[12,23,45,67,89,90]
y=[11,22,33,44,55,66]
x1=[9,98,87,76,65,54]
y1=[55,44,33,22,11,32]

plt.scatter(x,y,label="bar1 cahrtt",color="green",marker="<",s=150)
plt.scatter(x1,y1,label="bar2 cahrtt",color="red",marker="o",s=200)
plt.xlabel("xdata")
plt.title("bar cahrt")
plt.ylabel("ydata")
plt.legend()
plt.show()
```

```python
#to draw pi chart
import matplotlib.pyplot as plt
percent=[12,34,56,78,90]
item=["oppo",'micromax',"iphone",'mi',"oneplus"]
cols=['red','green','blue','pink','silver','yellow']
plt.pie(percent,
        labels=item,
       colors=cols,
       autopct="%1.1f%%",shadow=True)
       
plt.show()
```

```python
#import data from internet using link data in json format
import matplotlib.pyplot as plt
import requests
import json
#to requests json data from internet by link
a=requests.get("https://www.quandl.com/api/v3/datasets/NSE/21STCENMGM.json?api_key=FXeZpDiyM36xyyvtSjHm")
#load data from link in text format and store in 'b;
b=json.loads(a.text)
#to access specific set of data from json and store in 'xdata1' variable
data1=b['dataset']['data']
#to access specific of data from json and store in 'data2' variable
ydata=[i[1] for i in data1]
xdata=[i[0] for i in data1]
print(xdata)
print(ydata)

#to plot or visualise data in graph

plt.plot(xdata,ydata,label="sbi data",color='red')
plt.xlabel('date')
plt.ylabel('stock price')
plt.legend()
plt.title("sbi stock data")
plt.savefig('figure1.png')
plt.show()
```

```python
#to update figure in matplotlib with time
import matplotlib.pyplot as plt
import requests
import json
i=0
while(1):
    
    #to requests json data from internet by link
    a=requests.get("https://www.quandl.com/api/v3/datasets/NSE/21STCENMGM.json?api_key=FXeZpDiyM36xyyvtSjHm")
    #load data from link in text format and store in 'b;
    b=json.loads(a.text)
    #to access specific set of data from json and store in 'xdata1' variable
    data1=b['dataset']['data']
    #to access specific of data from json and store in 'data2' variable
    ydata=[i[1] for i in data1]
    xdata=[i[0] for i in data1]
    #print(xdata)
    #print(ydata)

    #to plot or visualise data in graph

    plt.plot(xdata,ydata,label="sbi data",color='red')
    plt.xlabel('date')
    plt.ylabel('stock price')
    plt.legend()
    plt.title("sbi stock data")
    plt.savefig('figure'+str(i)+'.png')
    i=i+1
    if i>5:
        break
```
